```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import load_boston
raw_data = load_boston()
raw_data                   # data not come in known format, need to process it
```

- We will try to generate the data by its given features & targets

## Preparing Dataset from Raw Data

```python
# Data Construction
import pandas as pd

df = pd.DataFrame(raw_data.data, columns=raw_data.feature_names)   # all features taken as columns
df['MEDV'] = raw_data.target                                       # Target Selected
df
```

- We got scaled data, so no need to apply scaler method e.g MinMax scaler

```python
# Finding Correlation
corr = df.corr()
corr                     # Returning correlation Matrix/ table from the range (-1 to + 1)
```

- Not so easy to understand, can apply heatmap for better understanding

# Heat Map

```python
import seaborn as sns

# HeatMap Application -1
plt.figure(figsize=(15,8))
import matplotlib.pyplot as plt
sns.heatmap(corr)
plt.show()
```

```python
# HeatMap Application -2
plt.figure(figsize=(15,8))
sns.heatmap(corr, annot=True)   # Annot True: showing corr values inside graph
plt.show()
```
