# OR Function / Pipe Operator

- Or/ pipe function returns when match any
- Condition: Find value for eihter "Rainy" OR visibility > 80

```python
(df['Weather']=='Rainy') | (df['Visibility_km']>80) 
```

```python
df[(df['Weather']=='Rainy') | (df['Visibility_km']>80)]  # Data Frame Conversion
```

```python
df[(df['city']=='Paris') | (df['Weather']=='Rainy')]  # Return either Paris or Rainy
```

## And Condtions

- Multiple conditons (as long as you want) can be added by using & operator
- All condtions must be matched or satisfy individually

```python
(df['Weather']=='Rainy') & (df['Rel Hum %']>30) & (df['Visibility_km']>25)  # Three condions applied
```

```python
df[(df['Weather']=='Rainy') & (df['Rel Hum %']>30) & (df['Visibility_km']>25)] # Data Frame
```
