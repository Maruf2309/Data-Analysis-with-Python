```python
data['Product'].value_counts().head(5).plot.bar()
plt.show()
```

```python
plt.figure(figsize=(30, 10))
data["Executive Name"].value_counts().plot.bar()
plt.title("Total Number of Transaction by Sales Offier", loc='left')
plt.show()
```

```python
plt.figure(figsize=(30, 10))
data["Sales Team"].value_counts().plot.bar()
plt.title("Total Number of Transaction by Sales Offier", loc='left')
plt.show()
```

```python
(data["Executive Name"].value_counts().head(10)/len(data)).plot.bar()
plt.title("Transaction Percentive by Sales Offier", loc='left')
plt.ylabel("Transaction Percentage")
plt.show()
```

```python
data['Executive Name'].value_counts().sort_index().plot.line()
plt.xticks(rotation='vertical', size=12)  # x axis showing in vertical axis
plt.show()
```

<p style ='font-family:Lucida Console; text-align:left; color: blue;font-size:18px'> Seaborn Countplot </p>

- sns countplot is very good alternaive of value count with a colorful approach

```python
sns.countplot(x='Product', data=data)  # sns is colurful
plt.show()
```

```python
y = data['Product'].value_counts()
mylabels = ["Furniture", "Office Supplies", "Technology","Stationary", "Fruits"]
plt.pie(y,labels = mylabels,startangle = 45 )
plt.show()
```

```python
y = data['Product'].value_counts()
mylabels = ["Furniture", "Office Supplies", "Technology","Stationary", "Fruits"]
myexplode = [0.2, 0, 0, 0,0]
plt.pie(y,labels = mylabels,explode = myexplode, shadow = True )
plt.show()
```

```python
y = data['Product'].value_counts()
mylabels = ["Furniture", "Office Supplies", "Technology","Stationary", "Fruits"]
mycolors = ["black", "hotpink", "b", "#4CAF50","purple"]
plt.pie(y,labels = mylabels,colors = mycolors )
plt.legend()
plt.show()
```

```python
data[data['Actual Sales'] > 500000]['Actual Sales'].plot.hist()
plt.show()
```

```python
data['Actual Sales'].plot.hist()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='Product', y='Actual Sales', data=data) # categorical value in x axis
plt.show()
```

```python
sns.barplot(x='Product',y='Collection', data=data, estimator=np.median)
```

```python
sns.barplot(x='Sales Team', y='Actual Sales', data=data, estimator=np.std)
plt.xticks(rotation='vertical', size=12)  # x axis showing in vertical axis
plt.show()
```

```python
sns.boxplot(x='Product', y='Actual Sales', data=data)  # x got categorical value
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(x='Sales Team', y='Actual Sales', data=data, hue='Product')
plt.xticks(rotation = 90)
plt.show()
```
